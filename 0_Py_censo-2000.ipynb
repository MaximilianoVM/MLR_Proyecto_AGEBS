```python
import csv
import pandas as pd

data = []
with open('DATA/censo-2000_Nacional.txt', 'r') as f:
    reader = csv.reader(f, delimiter=',')
    for row in reader:
        if len(row) == 23:  # Replace 23 with the expected number of fields
            data.append(row)

df = pd.DataFrame(data)
```

```python
#1st row is the header
df.columns = df.iloc[0]
df = df[1:]

df.head(1)
```

## tratamos datos irregulares y no deseados

```python
#tratamos * como NaN
df = df.replace('*', pd.NA)

#drop nan values
df = df.dropna()
```

* vamos a borrar los "TOTALES"

```python
#drop rows that in 1st, 3rd and 5th column include 'nacional', 'NACIONAL, 'Total', 'TOTAL', 'LOCALIDAD', 'LOCALIDADES'
df = df[~df['NOM_LOC'].str.contains('nacional|NACIONAL|Total|TOTAL|LOCALIDAD|LOCALIDADES')]

df.head(7)
```

```python
df['NOM_MUN'].unique()
```

```python
df[df['NOM_MUN'] == 'Aguascalientes'].head(7)
```

* vamos a borrar las filas donde el nombre del municipio sea el mismo que el de la localidad (creo que es un total)

```python
#drop rows where nom_mun == nom_loc
df = df[~(df['NOM_MUN'].str.upper() == df['NOM_LOC'].str.upper())]

df[df['NOM_MUN'] == 'Aguascalientes'].head(7)
```

### Numericos a INT64

```python
#only from column 4 to 22
df.iloc[:, 6:]
```

```python
columns_to_convert = ['P_TOTAL', 'POB0_4', 'POB18_', 'PNACOENT', 'GRADOESCO', 'PCASADA12_', 'P_1SM', 'P1_2SM', 'P2_5SM', 'P6_10SM', 'P10_SM', 'T_VIVHAB', 'VIVPARHAB', 'PRO_OCVP', 'PRO_OVP', 'VP_TV', 'TOTHOG']

def convert_to_float(x):
    try:
        return float(x)
    except ValueError:
        return None

df[columns_to_convert] = df[columns_to_convert].astype(str).apply(lambda col: col.map(convert_to_float))

df[columns_to_convert].dtypes
```

```python
df.describe()
```

```python
#muestra la localidad con mayor población
df[df['P_TOTAL'] == df['P_TOTAL'].max()]

#muestra la localidad con mayor 'P10_SM'
df[df['P10_SM'] == df['P10_SM'].max()]

#proporcion de la poblacion que tiene 10 o mas salarios minimos 
print(df['P10_SM'].sum() / df['P_TOTAL'].sum())

#proporcion de la poblacion que tiene 1 salario minimo 
print(df['P_1SM'].sum() / df['P_TOTAL'].sum())

#comparar proporciones de poblacion con P_1SM	P1_2SM	P2_5SM	P6_10SM	P10_SM	
df['P_1SM'].sum() / df['P_TOTAL'].sum(), df['P1_2SM'].sum() / df['P_TOTAL'].sum(), df['P2_5SM'].sum() / df['P_TOTAL'].sum(), df['P6_10SM'].sum() / df['P_TOTAL'].sum(), df['P10_SM'].sum() / df['P_TOTAL'].sum()
```
